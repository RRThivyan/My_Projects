# Simple Moving Average - SMA
Eg. for rolling(5) as mentioned above, we take a set of 5 values, say 'Open', and find the average of these 5 values and give them at the 5th position. Thus, we can see that the first 4 has Nan value and 5th has the average of all 5 values.

# Cumulative Moving Average - CMA
CMA, in a set of 'Open' column values, takes average value in a succession manner. for eg. the first value returns as Nan as its only one. second one is the average of first and second values. third is the average of first 3 values. 4th is the average of first 4 values. and so on.

```python
df['Open'].expanding().mean().plot(figsize=(10,5))
```

# EMA - Exponential Moving Average
formula : ((Open - prev EMA) * alpha) + prev EMA

first EMA = simple moving average, as we don't have any EMA initially

* alpha = 2/(span+1)
* span - rolling window
* for above eg. its 2/(5+1) = 2/6 = 0.3333

```python
df['EMA_0.1'] = df['Open'].ewm(alpha=0.1, adjust=False).mean()
```

```python
df['EMA_0.3'] = df['Open'].ewm(alpha=0.3, adjust=False).mean()
```

```python
df['EMA_0.5'] = df['Open'].ewm(alpha=0.5, adjust=False).mean()
```

```python
df[['Open', 'EMA_0.1', 'EMA_0.3', 'EMA_0.5']].plot(figsize=(15,5))
```

# EWMA - Exponential Weighted Moving Average

EWMA(t) = a * x(t) + (1-a) EWMA(t-1)

* a - weight
* EWMA(t-1) - previous EWMA value
* EWMA(t) - current EWMA value


```python
df['EWMA_5days'] = df['Open'].ewm(span=5).mean()
```

```python
df[['Open', 'EWMA_5days']].plot(figsize=(15,5))
```

# ARIMA
* AutoRegression - AR (PACF)
* Moving Average - MA (ACF)
* Integrated - I (Differencing)
* AR + MA - ARMA - Good for forecasting like Sales but bad for Stocks
* ACF/PACF - (Partial) Auto Correlation Function
* ARIMA - AR + I + MA

##### MA = mean + (error term * one coefficient)

Auto Regression 
- uses regression alogorithm
- old data is called lags
- K value - no of lags - calculated using PACF
- in correlation-lags plot, we have upper and lower bound. Correlations that are above these limits have high correlation and thus affecting the current value
- this plot is called PACF graph

ARIMA or SARIMAX
* good for stationary data

Stationary data - mean value, Standard Deviation are constant

1. Check the data for stationary

Hypothesis Dickey Fuller Test
* H0 >> unit root = 1 (Data not stationary)
* H1 >> unit root < 1 (Data is stationary)

1. AR -> P -> PACF
2. I -> d -> no of times differencing
3. MA -> q -> ACF

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def adf_test(series):
    result = adfuller(series)
    print('ADF Statistics: {}'.format(result[0]))
    print('p - value: {}'.format(result[1]))
    if result[1] <= 0.05:
        print('Strong evidence against the null hypothesis, reject null hypothesis. Data has no unit root and is stationary')
    else:
        print('Weak evidence against the null hypothesis, time series has a unit root, indicatiing it is a non stationary')
        
    
```

```python
adf_test(df1['#Passengers'])
```

```python
# Use Differencing techniques

df1['#Passengers 1st difference'] = df1['#Passengers'] - df1['#Passengers'].shift(1)
```

```python
df1.head()
```

```python
adf_test(df1['#Passengers 1st difference'].dropna())
```

```python
# Use Differencing techniques

df1['#Passengers 2nd difference'] = df1['#Passengers 1st difference'] - df1['#Passengers 1st difference'].shift(1)
```

```python
df1.head()
```

```python
adf_test(df1['#Passengers 2nd difference'].dropna())
```

```python
## 12 months

df1['#Passengers 12th difference'] = df1['#Passengers'] - df1['#Passengers'].shift(12)
```

```python
from datetime import datetime, timedelta
```

```python
train_dataset_end = datetime(1955, 12,1)
test_dataset_end = datetime(1960,12,1)
```

```python
train_data = df1[:train_dataset_end]
test_data = df1[train_dataset_end+timedelta(days=1):test_dataset_end]
```

```python
# Prediction
pred_start_date = test_data.index[0]
pred_end_date = test_data.index[-1]
```

```python
test_data
```

```python
# Create an ARIMA model
from statsmodels.tsa.arima.model import ARIMA
```

```python
train_data
```

```python
model_ARIMA = ARIMA(train_data['#Passengers'], order=(0,2,0))
```

```python
model_ARIMA_fit = model_ARIMA.fit()
```

```python
model_ARIMA_fit.summary()
```

```python
# Prediction
pred_start_date = test_data.index[0]
pred_end_date = test_data.index[-1]
print(pred_start_date)
print(pred_end_date)
```

```python
pred = model_ARIMA_fit.predict(start=pred_start_date, end=pred_end_date)
residuals = test_data['#Passengers'] - pred
```

```python
residuals
```

```python
model_ARIMA_fit.resid.plot(kind='kde')
```

```python
test_data['Predicted_ARIMA'] = pred
```

```python
test_data[['#Passengers', 'Predicted_ARIMA']].plot()
```

```python
## Create a SARIMAX model
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model_SARIMAX = SARIMAX(train_data['#Passengers'], order=(3,0,5), seasonal_order=(0,1,0,12))
```

```python
model_SARIMAX_fit = model_SARIMAX.fit()
```
