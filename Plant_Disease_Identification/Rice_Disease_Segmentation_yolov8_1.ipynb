```python
# total_white_pixel_count = 0
total_pixels_count = 0
total_white_pixel_perc = 0

for i in range(no_of_masks):
  bw_image = ((predict[0].masks.data[i].cpu().numpy() * 255).astype('uint8'))

  white_pixel_value = 255

  # Count the number of white pixels
  num_white_pixels = np.sum(bw_image == white_pixel_value)

  total_pixels = np.prod(bw_image.shape)

  # Calculate the percentage of white pixels
  percentage_white_pixels = (num_white_pixels / total_pixels) * 100
  total_white_pixel_perc += percentage_white_pixels

percentage_white_pixels = total_white_pixel_perc

print(f"Percentage of white pixels: {percentage_white_pixels:.2f}%")
```

```python
import os
image_folder_path = '/content/drive/MyDrive/AI_Projects/Rice/test_images'

image_files = [f for f in os.listdir(image_folder_path) if os.path.isfile(os.path.join(image_folder_path, f))]
```

```python
total_white_pixel_count = 0
total_pixels_count = 0
total_white_pixel_perc = 0

for image_file in image_files:
  image_path = os.path.join(image_folder_path, image_file)
  predict = model_pred.predict(image_path, save=True, save_txt=True)

  img = np.squeeze(predict[0].plot())
  rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  plt.imshow(rgb_image)
  plt.show()

  if isinstance(predict, list) and len(predict) > 0:
  # Extract the masks from the first element of the list
    masks = predict[0].masks

    # Check if masks is not None
    if masks is not None:
      # Continue with the original code...
      no_of_masks = masks.data.shape[0]

      for i in range(no_of_masks):
        bw_image = ((predict[0].masks.data[i].cpu().numpy() * 255).astype('uint8'))
        white_pixel_value = 255

        # Count the number of white pixels
        num_white_pixels = np.sum(bw_image == white_pixel_value)

        total_pixels = np.prod(bw_image.shape)

        # Calculate the percentage of white pixels
        percentage_white_pixels = (num_white_pixels / total_pixels) * 100
        total_white_pixel_perc += percentage_white_pixels


print(f"Percentage of Disease spread in the crop : {total_white_pixel_perc:.2f}%")
```
